# Part one: general¶

- Look at the activation_function and plot the y-values for each x from -5,5 spaced with 0.5

```python
import matplotlib.pyplot as plt
import numpy as np

def activation_function(x):
    if x < 0:
        return -1
    else:
        return 1
    
rnge = np.linspace(-5, 5, num=21)
values = [activation_function(i) for i in rnge]
plt.plot(rnge, values)
plt.axis([-5, 5, -2, 2])

```

- Change the perceptron method from the notebook to use the numpy.dot() method in line 12 instead of the lengthy sum() function

```python
def perceptron(inp, weights):
    dot_product = np.dot(inp, weights)
    output = activation_function(dot_product)
    return output

perceptron([1, 2, 3, 4, 5], [1, 1, 2, 1, 1])
```

```python
from sklearn import preprocessing
import pandas as pd

data = pd.read_csv('ratmouse.csv')
label_enc = preprocessing.LabelEncoder()
data['type'] = label_enc.fit_transform(data['type'].astype(str))

data.plot.scatter(x='weight', y='height', c='type', colormap='viridis')
slope, intercept = np.polyfit(data['weight'], data['height'], 1)

slope = -1/slope
intercept = 1179.7
plt.plot(data['weight'], slope*data['weight'] + intercept, 'b')

plt.show()
```
